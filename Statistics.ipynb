```python
Laliga_df = pd.read_csv("C:\\Users\\KRISH\\Desktop\\Statistics\\Laliga.csv")
```

```python
Lal_df = Laliga_df.replace('-', np.nan)
```

```python
Lal_df.fillna(0, inplace=True)
```

```python
new_header = Lal_df.iloc[0] 
Lal_df = Lal_df[1:] 
Lal_df.columns = new_header
Lal_df.head()
```

```python
def Goal_diff_count(GoalsFor,GoalsAgainst):
    Goal_diff_count = GoalsFor - GoalsAgainst
    #print ("Goal difference is : ", Goal_diff_count)
    return Goal_diff_count
```

```python
Lal_df['Goal_diff_count'] = Goal_diff_count(Lal_df['GoalsFor'].astype(int),Lal_df['GoalsAgainst'].astype(int))
Lal_df['Goal_diff_count']
```

```python
Lal_df['Goal_diff_count'] = Goal_diff_count(Lal_df['GoalsFor'].astype(int),Lal_df['GoalsAgainst'].astype(int))
Lal_df.sort_values('Goal_diff_count', ascending = False)[['Goal_diff_count','Team','Points']].head(10)
```

```python
Lal_df['Goal_diff_count'].min()
```
